```python
import pandas as pd
import re
po_value = input("Enter the PO value (e.g., PO-106070): ")
item_no = input("Enter the Item No value (e.g., COLVNM04Y): ")
df = pd.read_excel(r'C:\Users\Admin\Downloads\pdf2excel.xlsx', skiprows=2) ## Read the Excel file starting from row 11 (skip first 10 rows)
# Define the column names exactly as they appear, including newline characters
selected_columns = ['Serial\nNo', 'Description', 'Stamp', 'Pieces']

# Select only the required columns
df_selected = df[selected_columns].copy()

# Clean the 'Description' column by replacing newlines with spaces
df_selected['Description'] = df_selected['Description'].str.replace('\n', ' ', regex=True)
df_selected['Stamp'] = df_selected['Stamp'].str.replace('\n', ' ', regex=True)
# Rename columns
df_selected.rename(columns={
    'Serial\nNo':'SerialNo',
    'Description': 'CustomerProductionInstruction',
    'Pieces':'OrderItemPcs'
    
}, inplace=True)
df_selected = df_selected[~df_selected['SerialNo'].isin(['Buyer', 'Serial\nNo'])]

df_selected.insert(0, 'SrNo', range(1, len(df_selected) + 1))

SrNo_index = df_selected.columns.get_loc('SrNo')
df_selected.insert(loc=SrNo_index + 1, column='StyleCode', value='')
df_selected.insert(loc=SrNo_index + 2, column='ItemSize', value='')
df_selected.insert(loc=SrNo_index + 3, column='OrderQty', value='10')

# Move 'OrderItemPcs' to right of 'OrderQty'
orderqty_index = df_selected.columns.get_loc('OrderQty')
orderitempcs_col = df_selected.pop('OrderItemPcs')
df_selected.insert(loc=orderqty_index + 1, column='OrderItemPcs', value=orderitempcs_col)

OrderItemPcs_index = df_selected.columns.get_loc('OrderItemPcs')
df_selected.insert(loc=OrderItemPcs_index + 1, column='Metal', value='')
def extract_metal(text):
    if pd.notna(text):
        match = re.search(r'14KY', text)
        if match:
            return 'G14Y'
    return ''

Metal_index = df_selected.columns.get_loc('Metal')
df_selected.insert(loc=Metal_index + 1, column='Tone', value='')
df_selected['Tone'] = df_selected['CustomerProductionInstruction'].apply(
    lambda x: 'YG' if pd.notna(x) and '14KY' in x.upper() else ''
)
df_selected['Metal'] = df_selected['CustomerProductionInstruction'].apply(extract_metal)

Tone_index = df_selected.columns.get_loc('Tone')
df_selected.insert(loc=Tone_index + 1, column='ItemPoNo', value=po_value)

itempono_index = df_selected.columns.get_loc('ItemPoNo')
# Insert 'ItemRefNo' after 'ItemPoNo','ItemRefNo','StockType'
df_selected.insert(loc=itempono_index + 1, column='ItemRefNo', value='')
df_selected.insert(loc=itempono_index + 2, column='StockType', value='')
df_selected.insert(loc=itempono_index + 3, column='MakeType', value='')

CustomerProductionInstruction_index = df_selected.columns.get_loc('CustomerProductionInstruction')
df_selected.insert(loc=CustomerProductionInstruction_index + 1, column='SpecialRemarks', value='')
df_selected.insert(loc=CustomerProductionInstruction_index + 2, column='DesignProductionInstruction', value='')
df_selected.insert(loc=CustomerProductionInstruction_index + 3, column='StampInstruction', value='')

Stamp_index = df_selected.columns.get_loc('Stamp')
df_selected.insert(loc=Stamp_index + 1, column='OrderGroup', value='')
df_selected.insert(loc=Stamp_index + 2, column='Certificate', value='')
df_selected.insert(loc=Stamp_index + 3, column='SKUNo', value=item_no)

# List of columns to insert after 'SKUNo'
new_columns = [
    'Basestoneminwt', 'Basestonemaxwt', 'Basemetalminwt', 'Basemetalmaxwt',
    'Productiondeliverydate', 'Expecteddeliverydate', 'null', 'SetPrice', 'StoneQuality'
]

# Find the index of 'SKUNo'
sku_index = df_selected.columns.get_loc('SKUNo')

# Insert each column at the correct position
for i, col_name in enumerate(new_columns):
    df_selected.insert(loc=sku_index + 1 + i, column=col_name, value='')
# Generate StyleCode
def generate_style_code(row):
    if pd.notna(row['CustomerProductionInstruction']) and '18IN' in row['CustomerProductionInstruction']:
        tone = row['Tone'] if pd.notna(row['Tone']) else ''
        sku = row['SKUNo'] if pd.notna(row['SKUNo']) else ''
        sku_suffix = 'CO' if 'CO' in sku else ''
        return f"XK2807G-18IN{tone}{sku_suffix}"
    return ''

df_selected['StyleCode'] = df_selected.apply(generate_style_code, axis=1)

# Generate SpecialRemarks
def generate_special_remarks(row):
    remarks = []
    sku = row['SKUNo']
    desc = row['CustomerProductionInstruction']

    if pd.notna(sku):
        remarks.append(sku)
    if pd.notna(desc) and '14KY' in desc:
        remarks.append('14K YELLOW GOLD')
    if pd.notna(desc) and '18IN' in desc:
        remarks.append('SZ 18 INCH')
    
    # Always append diamond quality
    remarks.append('DIA QLTY-HI-VS')
    
    return ','.join(remarks)

df_selected['SpecialRemarks'] = df_selected.apply(generate_special_remarks, axis=1)

# ✅ Generate StampInstruction row-wise using SrNo
def generate_stamp_instructions(df, base_serial_start=301133):
    stamp_instructions = []
    for idx, row in df.iterrows():
        srno = row['SrNo']
        start_serial = base_serial_start + (srno - 1) * 10
        end_serial = start_serial + 9

        # Condition checks
        has_ufjc = 'UFJC' in str(row['Stamp'])
        has_14ky = '14KY' in str(row['CustomerProductionInstruction'])
        qty_is_10 = str(row['OrderQty']) == '10'

        # Find CTW like "0.70CTW"
        ctw_match = re.search(r'\d+\.\d+CTW', str(row['CustomerProductionInstruction']))
        ctw_value = ctw_match.group() if ctw_match else ''

        if has_ufjc and has_14ky and qty_is_10 and ctw_value:
            instruction = f"UFJC 14KY {start_serial} to {end_serial} {ctw_value}"
        else:
            instruction = ''

        stamp_instructions.append(instruction)
    return stamp_instructions

# Assign StampInstruction
df_selected['StampInstruction'] = generate_stamp_instructions(df_selected)

# Show first 20 rows
df_selected.head(500)
df_selected.to_excel("GATI_FORMAT_UNEEK.xlsx", index=False)
```
